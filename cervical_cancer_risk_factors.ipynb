```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
risk = pd.read_csv('risk_factors_cervical_cancer.csv')
```

```python
risk
```

```python
sns.boxplot(x='Dx:Cancer', y='Dx:HPV', data=risk)
```

```python
risk.corr().style.background_gradient(cmap="GnBu")
```

```python
g = sns.FacetGrid(risk, col = 'Hormonal Contraceptives', row = 'IUD')
g = g.map(plt.hist, 'Dx:Cancer')
```

```python
sns.scatterplot(x='Dx:Cancer', y='Smokes', data=risk)
```

```python
sns.scatterplot(x='Dx:HPV', y='Age', data=risk)
```

```python
sns.swarmplot(x='Dx:HPV', y='Age', data=risk, size = 3)
```

```python
sns.stripplot(x='Dx:HPV', y='Age', data=risk)
```

```python
my_palette = {0: 'green', 1:'red'}
sns.boxplot(x='Dx:Cancer', y='Age', data=risk, palette=my_palette)
sns.swarmplot(x='Dx:Cancer', y='Age', data=risk, color = 'black', size = 3)
```

```python
plt.matshow(risk.corr())
```

```python
sns.pointplot(x='Smokes (years)', y='Dx:Cancer', data=risk)
```

```python
sns.barplot(x ='STDs (number)', y = 'Dx:Cancer', data=risk)
sns.swarmplot(x = 'STDs (number)', y = 'Dx:Cancer', data = risk, size = 3)
```

```python
sns.stripplot(x = 'STDs (number)', y = 'Dx:Cancer', data = risk)
```

```python
g = sns.FacetGrid(risk, col = 'STDs (number)', row = 'STDs')
g = g.map(plt.hist, 'Dx:Cancer').add_legend()
```

```python
sns.pairplot(hue = 'Dx:Cancer', data = risk)
```

```python
sns.boxplot(x = 'Number of sexual partners', y = 'Dx:Cancer', data = risk)
```

```python
risk_sorted = risk.sort_values('Number of sexual partners', ascending = True)
sns.pairplot(y_vars = ['Number of sexual partners', 'First sexual intercourse'], x_vars = ['Dx:Cancer', 'Dx:HPV'], data = risk_sorted)
```
